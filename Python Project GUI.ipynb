```python
import tkinter
from tkinter import PhotoImage, Label, Button, Toplevel, StringVar, Entry, Menu, END
from tkinter import filedialog
import os
import string
import re
import pytesseract
import cv2
from PIL import ImageTk,Image
from PyDictionary import PyDictionary
from gtts import gTTS
from email_validator import validate_email, EmailNotValidError

#Global variables used for certain matters such as page indentation
Check = 3
UCount = 0
LCount = 0
HCount = 0
MCount = 1
Trial = 0

try:
    class MandP:
        def __init__(self, Word, menu_screen, user):
            self.word = Word
            self.buttons = Button(menu_screen, text=self.word, justify="left", width=10, bd=0, cursor="mouse", command = lambda:self.DictionaryWork(self.word, menu_screen, user))
            self.buttons.pack()

        # Getting meanings and pronunciation
        def DictionaryWork(self, find, menu_screen, user):
            mean_screen = Toplevel(menu_screen)
            mean_screen.geometry("700x700")
            dictionary=PyDictionary('features="lxml"')
            Meanings = dictionary.meaning(find)
            Label(mean_screen,text=find, font=('Verdana', 16, 'bold italic'), bd=0, width=700).pack()
            Label(mean_screen,text="", font=('Verdana', 16, 'bold italic'), bd=2, width=700).pack()
            obj1 = gTTS(text=find,lang='en',slow=False)
            obj1.save('E:/Audio.mp3')
            Button(mean_screen, text="Pronunciation", font=('Verdana',10,'bold italic'), bd=2, bg="black", fg="snow", width=20, command = lambda:os.system('start E:/Audio.mp3')).pack()
            for i,j in Meanings.items():    
                mcount = 1
                Label(mean_screen, text=i, font=('Verdana', 13, 'italic'), bd = 0, justify="left", width = 700).pack()
                for k in j:
                    Label(mean_screen, text=(str(mcount)+". "+k), font=('Verdana',10), bd=0, justify="left", width=700).pack()
                    mcount+=1
                Label(mean_screen, text="", bd=0, width=500).pack()
            try:
                fobj = open((user+" History.txt"),"a+")
                if (find+"\n") not in fobj.readlines():
                    fobj.write(find+"\n") 
                fobj.close()
            except Exception as ex:
                print("Exception in file handling around line 50.", ex)

    def BacktoMainScreen():
        global main_screen
        global register_screen
        global login_screen
        global menu_screen
        global Check

        if Check != 3:
            main_screen.deiconify()
        if Check == 1:
            register_screen.destroy()
            Check = 3
        elif Check == 0:
            login_screen.destroy()
            Check = 3
        elif Check == 2:
            menu_screen.destroy()
            Check = 3


    def ValidateEmail(email):
        try:
            v = validate_email(email) 
            email = v["email"] 
            return True
        except EmailNotValidError as e:
            #print(str(e))
            return False

    def ProperUsername(username):
            if ((username[0] in string.ascii_uppercase) and (len(username)>=4)):
                return True
            else:
                return False


    def Quit():
        global main_screen
        main_screen.destroy()

    #Designing Registration Window
    def register():
        global register_screen
        global logo
        register_screen = Toplevel(main_screen)
        main_screen.withdraw()
        register_screen.title("Register")
        register_screen.geometry("1345x730")
        global username
        global password
        global eid
        global re_password
        global Check
        username = StringVar()
        password = StringVar()
        eid = StringVar()
        re_password = StringVar()

        register_screen.configure(background="spring green")

        Check = 1
        Button(register_screen, text="Back to Main Screen", fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command=BacktoMainScreen).pack(pady=10)
        Button(register_screen, text="Quit", fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command=Quit).pack(pady=10)
        for i in range(2):
            Label(register_screen, text="",bg="spring green").pack()
        Label(register_screen,image=logo,bg="maroon",bd=4).pack(side="top")
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="Please enter registration details below", bg="DarkOrchid3", width=30, height=1, bd=3, fg="snow", font=('Helvetica',12,'italic underline')).pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="Email ID: * ", bd=3, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Entry(register_screen, textvariable=eid).pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="Username: * ", bd=3, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Entry(register_screen, textvariable=username).pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="Password: * ", bd=3, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Entry(register_screen, textvariable=password, show='*').pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="Confirm Password: * ", bd=3, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Entry(register_screen, textvariable=re_password, show='*').pack()
        Label(register_screen, text="", bg="spring green").pack()
        Label(register_screen, text="", bg="spring green").pack()
        Button(register_screen, text="Register", width=12, height=1, bd=1, fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command = register_user).pack()

    # Designing Login window 
    def login():
        global login_screen
        login_screen = Toplevel(main_screen)
        main_screen.withdraw()
        login_screen.title("Login")
        login_screen.geometry("1345x730")
        login_screen.configure(background="spring green")
        global username_verify
        global password_verify
        global Check

        username_verify = StringVar()
        password_verify = StringVar()
        Check = 0
        Button(login_screen, text="Back to Main Screen", fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command=BacktoMainScreen).pack(pady=10)
        Button(login_screen, text="Quit", fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command=Quit).pack(pady=10)
        for i in range(2):
            Label(login_screen, text="",bg="spring green").pack()
        Label(login_screen,image=logo,bg="maroon",bd=4).pack(side="top")
        Label(login_screen, text="", bg="spring green").pack()
        Label(login_screen, text="", bg="spring green").pack()
        Label(login_screen, text="Please enter login details below", bg="DarkOrchid3", width=30, height=1, bd=3, fg="snow", font=('Helvetica',12,'italic underline')).pack()
        Label(login_screen, text="", bg="spring green").pack()
        Label(login_screen, text="Username: * ", bd=2, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Label(login_screen, text="",bg="spring green").pack()
        Entry(login_screen, textvariable=username_verify).pack()
        Label(login_screen, text="",bg="spring green").pack()
        Label(login_screen, text="Password: * ", bd=2, bg="spring green", font=('Verdana',10,'bold'), fg="black").pack()
        Label(login_screen, text="",bg="spring green").pack()
        Entry(login_screen, textvariable=password_verify, show= '*').pack()
        Label(login_screen, text="", bg="spring green").pack()
        Label(login_screen, text="", bg="spring green").pack()
        Button(login_screen, text="Login", width=12, height=1, bd=1, fg="snow", cursor="mouse", font=('system', 10, 'italic'), bg="midnightblue", highlightcolor="black", highlightbackground="black", command = login_verify).pack()

    # Implementing event on register button - storing the registration information if valid
    def register_user():
        global UCount
        global button1
        global button2
        global button3
        global Trial
        global T1
        global T2
        global T3
        global CurrentUser

        T1 = Label(register_screen, text="", bg="spring green")
        T2 = Label(register_screen, text="", bg="spring green")
        T3 = Label(register_screen, text="", bg="spring green")
        if UCount>0:
            button1.invoke()
            button2.invoke()
            button3.invoke()
            button4.invoke()
        eid_info = eid.get()
        username_info = username.get()
        password_info = password.get()
        cpass_info = re_password.get()
        if ((username_info+".txt") not in os.listdir()):
            if (password_info == cpass_info):
                try:
                    T2.pack()
                    if ValidateEmail(eid_info):
                        if ProperUsername(username_info): 
                                File = open("EUName.txt","a")
                                File.write(eid_info+"\t\t\t"+username_info)
                                File.close()
                                username_info = username_info+".txt"
                                File = open(username_info, "w")
                                File.write(password_info)
                                File.write("\n")
                                File.close()
                                Entry(register_screen, textvariable=eid).delete(0,END)
                                Entry(register_screen, textvariable=username).delete(0, END)
                                Entry(register_screen, textvariable=password).delete(0, END)
                                Entry(register_screen, textvariable=re_password).delete(0,END)

                                Trial = Label(register_screen, text="Registration Success!", bg="spring green", fg="purple", font=("Verdana", 11,'italic'))
                                Trial.pack()
                                T3.pack()
                                CurrentUser = username_info
                                Check = 1
                                MenuWindow()
                        else:
                            Entry(register_screen, textvariable=username).delete(0, END)
                            UCount+=1
                            T1.pack()
                            Trial = Label(register_screen,text="Requirement: first character of username between A-Z, minimum length: 4 characters.",bg="spring green", fg="purple")
                            Trial.pack()

                    else:
                        Entry(register_screen, textvariable=eid).delete(0,END)
                        Trial = Label(register_screen, text="Invalid Email ID", bg="spring green", fg="purple", font=("Verdana", 11,'italic'))
                        Trial.pack()

                except Exception as e1:
                    print("Exception encountered.",e1)

            elif (password_info!=cpass_info):
                T2.pack()
                Entry(register_screen, textvariable=password).delete(0, END)
                Entry(register_screen, textvariable=re_password).delete(0,END)    
                Trial = Label(register_screen,text="Registration failed!\nPasswords do not match.")
                Trial.pack()
            UCount = UCount+1
        else:
            T2.pack()
            Entry(register_screen, textvariable=username).delete(0,END)
            Trial = Label(register_screen, text="Registration Failed.\nUsername is already in use.")
            Trial.pack()
        button1 = Button(fg="spring green", bg="spring green", command=lambda:Trial.pack_forget())
        button4 = Button(fg="spring green", bg="spring green", command=lambda:T1.pack_forget())
        button2 = Button(fg="spring green", bg="spring green", command=lambda:T2.pack_forget())
        button3 = Button(fg="spring green", bg="spring green", command=lambda:T3.pack_forget())

    # Implementing event on login button - checking for matching username and password
    def login_verify():
        global T2
        global T3
        global Trial
        global LCount
        global Check
        global CurrentUser

        username1 = username_verify.get()+".txt"
        password1 = password_verify.get()
        T2 = Label(login_screen, text="", bg="spring green")
        T3 = Label(login_screen, text="", bg="spring green")
        if LCount>0:
            Trial.destroy()
            T2.destroy()
            T3.destroy()
        list_of_files = os.listdir()
        if username1 in list_of_files:
            file1 = open(username1, "r")
            verify = file1.readline()
            if (password1+"\n") == verify:
                CurrentUser = username1
                Check = 0
                MenuWindow()
            else:
                T2.pack()
                Entry(login_screen, textvariable=password_verify).delete(0,END)
                Trial = Label(login_screen, text="Password Incorrect.\nEnter correct password.")
                Trial.pack()
                Entry(login_screen, textvariable=password_verify).delete(0, END)
                LCount+=1

        else:
            T3.pack()
            Entry(login_screen, textvariable=username_verify).delete(0,END)
            Trial = Label(login_screen, text="Invalid Username.\nCheck validity of username.")
            Trial.pack()
            LCount+=1

    # A part of the Menu Screen
    def History(fname):
        global menu_screen
        global HCount
        global MCount
        global HL
        global HB
        global LD
        global Breaks
        global BD
        global Trial

        for i in LD:
            i.destroy()
        for i in Breaks:
            i.buttons.destroy()
        BD.destroy()
        if HCount == 0:
            HL = list([])
            HCount+=1
        else:
            for i in HL:
                i.destroy()
            if MCount==0 and Trial==0:
                HB.destroy()
                Trial = 1
                MCount = 1
            HCount=0

        fname = fname+" History.txt"
        try:
            count = 1
            if fname in os.listdir():
                fobj = open(fname,"r")
                j=0
                print(fname)
                for i in fobj:
                    print(i)
                    HL.append(Label(menu_screen, text=(str(count)+". "+i), width=1345, bd=0, justify="left"))
                    HL[j].pack()
                    count+=1
                    j+=1
                fobj.close()
            else:
                HB = Label(menu_screen, text="User hasn't requested for any word yet.", justify="center", font=('Verdana', 13, 'italic'))
                HB.pack()
                Trial = 0
                MCount = 0
        except Exception as ex:
            print("Check for file errors or screen errors around line 320.", ex)

    #Implementation of File Upload Button - selection of jgp file for use
    def FileUpload():
        global menu_screen
        global LD
        global BD
        global Breaks
        global CurrentUser

        Words=[]
        menu_screen.filename = filedialog.askopenfilename(initialdir = "C:/Users/Vostro",title = "Select JPEG or PNG file",filetypes = (("jpeg files","*.jpg"),("png files","*.png")))
        for i in LD:
            i.destroy()
        BD.destroy()  
        LD = []
        try:
            Images = cv2.imread(menu_screen.filename)
            TextObtained = pytesseract.image_to_string(Images)
            TextObtained = TextObtained.replace('?',' ').replace('.',' ').replace(',',' ').replace('\n',' ').replace(':',' ').replace(';',' ')
            TextObtained = TextObtained.replace('/',' ').replace('!',' ').replace('*',' ').replace('&',' ').replace('\'',' ').replace('\"',' ')
            TextObtained = TextObtained.replace('(',' ').replace(')',' ').replace('{',' ').replace('}',' ').replace('[',' ').replace(']',' ')
            TextObtained = TextObtained.replace('\\',' ').replace('-',' ')
            TextObtained = re.sub(' +', ' ', TextObtained)
            Words = re.split(' ',TextObtained)
            LD.append(Label(menu_screen, text=" ", width=1345))
            LD[0].pack()
            i=0
            Breaks = []
            for i in range(len(Words)):
                temp = MandP(Words[i], menu_screen, CurrentUser.strip(".txt"))
                Breaks.append(temp)
        except Exception as ex:
            print(ex)

    #Main Menu Screen
    def MenuWindow():
        global Check
        global LD
        global BD
        global HB
        global HL
        global HCount
        global Trial
        global Breaks
        global login_screen
        global main_screen
        global register_screen
        global MCount
        global CurrentUser
        global menu_screen

        Breaks = []
        if Check == 0:
            login_screen.withdraw()
        elif Check == 1:
            register_screen.withdraw()
        if Check != 2:
            Check = 2
            LD = []
            menu_screen = Toplevel(main_screen)
            menu_screen.title("User: "+CurrentUser.strip(".txt"))
            menu_screen.geometry("1345x730")
            Breaks.append(MandP("none", menu_screen,CurrentUser.strip(".txt")))
            Label(menu_screen, text="Image-ry Dictionary", bd=0, width=1345, height=4, justify="center", font=('Verdana',18,'bold italic')).pack()
            Label(menu_screen, text="", bd=1, width=1345).pack()
            Label(menu_screen, text="", bd=0, width=8).pack(padx=0)
        if HCount>0:
            HCount = 0
            if Trial!=1 and MCount==0:
                HB.destroy()
                MCount = 1
                Trial = 1
            for i in HL:
                i.destroy()
        BD = Button(menu_screen, text="Upload JPEG/PNG Files", highlightcolor="black", highlightbackground="black", highlightthickness=5, width=33, height=2, bd=5, bg="DarkOrchid4", fg="white", relief="raised", command = FileUpload)
        Menubar = Menu(menu_screen)
        Menubar.add_command(label="Home", command = MenuWindow)
        Menubar.add_command(label="Word History", command = lambda:History(CurrentUser.strip(".txt")))
        Menubar.add_command(label="Logout", command = BacktoMainScreen)
        Menubar.add_command(label="Quit", command=Quit)
        menu_screen.configure(menu=Menubar)
        for i in LD:
            i.destroy()
        for i in Breaks:
            i.buttons.destroy()
        LD = []
        for i in range(8):
            LD.append(Label(menu_screen, text=""))
            LD[i].pack()
        BD.pack()


    #Main Login/Register Screen
    def main_account_screen():
        global main_screen
        global logo
        main_screen = tkinter.Tk()
        main_screen.geometry("1345x730")
        main_screen.title("Account Login")
        main_screen.configure(background="spring green")
        logo = PhotoImage(file="nirma.gif")
        for i in range(8):
            Label(text="",bg="spring green").pack()
        Label(main_screen,image=logo,bg="maroon",bd=4).pack(side="top")
        Label(text="", bg="spring green").pack()
        Label(text="",bg="spring green").pack()
        Label(text="Login/Register", bg="DarkOrchid4", width="30", height="2", bd=4, font=("Calibri", 13),fg="white",relief="ridge").pack()
        Label(text="",bg="spring green").pack()
        Button(text="Login", bg="midnightblue", highlightcolor="black", highlightbackground="black", highlightthickness=5, bd=5, height="2", width="30", fg="snow", relief="raised", cursor="mouse", command = login).pack()
        Label(text="",bg="spring green").pack()
        Button(text="Register", bg="midnightblue", highlightcolor="black", highlightbackground="black", highlightthickness=5, bd=5, height="2", width="30", fg="snow", relief="raised", cursor="mouse", command=register).pack()    
        main_screen.mainloop()
    main_account_screen()

except Exception as ex:
    print(ex)
```
